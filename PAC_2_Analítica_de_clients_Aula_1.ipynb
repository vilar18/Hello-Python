####b. Té el dataset sesgues de gènere?

El dataset conté dades de 7994 clients, amb una distribució de gènere lleugerament desequilibrada: **55.2% de les observacions corresponen a dones (FEMALE) i 44.8% a homes (MALE)**. Aquesta diferència d'aproximadament 10 punts percentuals indica una sobrerepresentació moderada de dones en la mostra. Tot i que no és un desequilibri extrem, és un factor a tenir en compte, ja que models entrenats amb aquestes dades podrien, potencialment, generalitzar lleugerament millor per al grup majoritari si no s'apliquen tècniques de mitigació.

Les mètriques numèriques clau no presenten diferències significatives que suggereixin un biaix sistèmic clar en les variables financeres principals. Per exemple, el SALDO_CUENTA mitjà i la mediana són molt similars entre gèneres (mitjana: 1566 vs 1521; mediana: 888 vs 842). El mateix passa amb l'IMPORTE_COMPRAS (mitjana: 999 vs 957; mediana: 360 vs 349) i, de forma destacada, amb el LIMITE_CREDITO, on la mitjana és gairebé idèntica (4484 vs 4460) i la mediana és exactament la mateixa (3000) per a ambdós grups. Això suggereix que, **almenys a nivell central de la distribució, no hi ha una diferència significativa en el límit de crèdit assignat** basat únicament en el gènere registrat.

Les lleugeres diferències observades en IMPORTE_PAGOS, TOTAL_VISITAS_ONLINE i NUM_TRANSACCIONES_COMPRAS (amb valors lleugerament superiors per a les dones) no semblen prou grans per indicar un biaix.

Amb tot això, l'anàlisi de la variable JOB_SECTOR revela una segregació de gènere molt marcada. Aquesta és l'evidència més clara d'un patró relacionat amb el gènere dins del dataset. Observem que sectors com Construcció (0% dones vs 12.6% homes), Enginyeria (3.2% vs 17.5%), Finances (4.8% vs 12.5%) i especialment Tecnologia (5.2% vs 32.6%) estan fortament dominats per homes en aquesta mostra. Contràriament, Màrqueting (37.5% dones vs 0% homes), Retail (20.5% vs 5.2%) i Sanitat (19.5% vs 10.8%) tenen una representació femenina molt més alta. Aquesta forta correlació entre gènere i sector laboral implica **que JOB_SECTOR actua com un proxy potencialment fort per a GENDER**. Si aquesta variable s'utilitza en models predictius, podria introduir o amplificar biaixos de gènere indirectament, fins i tot si el gènere mateix no s'utilitza explícitament.

Finalment, l'anàlisi dels valors nuls en LIMITE_CREDITO i IMPORTE_MINIMO_PAGOS mostra percentatges de dades faltants molt similars entre homes i dones (0% vs 0.03% per al límit de crèdit; 3.47% vs 3.63% per als pagaments mínims). Això suggereix que la **qualitat de les dades en termes d'exhaustivitat** no sembla estar esbiaixada per gènere en aquestes columnes específiques.

En **conclusió**, tot i que les mètriques financeres directes com el límit de crèdit no mostren biaixos alarmants entre gèneres, el dataset presenta un desequilibri moderat en la representació i, de manera molt significativa, una forta segregació per gènere en els sectors laborals. Aquest últim punt és el principal focus de preocupació pel que fa al biaix potencial.



```python
# 1. Distribució de la variable Gènere
print("\n1. Distribució de Gènere:")
gender_counts = df['GENDER'].value_counts()
gender_percentages = df['GENDER'].value_counts(normalize=True) * 100
print("Recompte per Gènere:\n", gender_counts)
print("\nPercentatge per Gènere:\n", gender_percentages)

# Visualització amb Plotly Express
fig_gender_dist = px.bar(
    x=gender_counts.index,
    y=gender_counts.values,
    labels={'x': 'Gènere', 'y': 'Nombre de Clients'},
    title='Distribució de Gènere en el Dataset',
    text_auto=True, # Mostra els valors a sobre de les barres
    color_discrete_sequence=px.colors.qualitative.Pastel # Esquema de colors
)
fig_gender_dist.update_layout(xaxis_title="Gènere", yaxis_title="Nombre de Clients")
fig_gender_dist.show() # Mostra el gràfic

# 2. Comparació de mètriques numèriques clau per gènere
print("\n\n2. Comparació de Mètriques Numèriques Mitjanes per Gènere:")
numeric_cols_to_compare = [
    'SALDO_CUENTA', 'IMPORTE_COMPRAS', 'LIMITE_CREDITO',
    'IMPORTE_PAGOS', 'TOTAL_VISITAS_ONLINE', 'NUM_TRANSACCIONES_COMPRAS'
]
gender_comparison = df.groupby('GENDER')[numeric_cols_to_compare].agg(['mean', 'median'])
print(gender_comparison)

# Visualització comparativa (Boxplots en subplots amb Plotly)
# Variables a visualitzar als boxplots
boxplot_vars = ['SALDO_CUENTA', 'IMPORTE_COMPRAS', 'LIMITE_CREDITO']
# Títols per als subplots
subplot_titles = [f'Comparativa {var.replace("_", " ").title()} per Gènere' for var in boxplot_vars]

# Crear figura amb subplots
fig_boxplots = make_subplots(rows=1, cols=len(boxplot_vars), subplot_titles=subplot_titles)

# Afegir traces de boxplot a cada subplot
for i, var in enumerate(boxplot_vars):
    # Afegim una traça per cada gènere per poder assignar colors diferents si volem
    # O simplement deixem que Plotly ho gestioni amb x='GENDER'
    fig_boxplots.add_trace(
        go.Box(y=df[var], x=df['GENDER'], name=var, marker_color=px.colors.qualitative.Pastel[i]), # Assigna un color per variable
        row=1, col=i+1
    )
    # Actualitzem els eixos per aquest subplot específic
    fig_boxplots.update_yaxes(title_text=var.replace("_", " ").title(), row=1, col=i+1)
    fig_boxplots.update_xaxes(title_text="Gènere", row=1, col=i+1)


# Actualitzar el layout general
fig_boxplots.update_layout(
    title_text='Comparativa de Distribucions Numèriques per Gènere',
    height=500, # Ajusta l'alçada si cal
    showlegend=False # La llegenda no és molt útil aquí
)
fig_boxplots.show() # Mostra el gràfic combinat

# 3. Relació entre Gènere i altres variables categòriques (ex: JOB_SECTOR)
print("\n\n3. Relació entre Gènere i Sector Laboral (JOB_SECTOR):")
cross_tab = pd.crosstab(df['GENDER'], df['JOB_SECTOR'])
print("\nTaula de Contingència (Recomptes):\n", cross_tab)
cross_tab_norm = pd.crosstab(df['GENDER'], df['JOB_SECTOR'], normalize='index') * 100
print("\nTaula de Contingència (Percentatge per Gènere):\n", cross_tab_norm)

# Visualització amb Plotly Express (Grouped Bar Chart usant histogram)
fig_job_gender = px.histogram(
    df,
    x='JOB_SECTOR',
    color='GENDER',  # Agrupa i coloreja per gènere
    barmode='group',
    labels={'count': 'Nombre de Clients', 'JOB_SECTOR': 'Sector Laboral', 'GENDER': 'Gènere'},
    title='Distribució de Sectors Laborals per Gènere',
    category_orders={"JOB_SECTOR": sorted(df['JOB_SECTOR'].unique())}, # Opcional: ordena els sectors
    color_discrete_map={'MALE': px.colors.qualitative.Pastel[0], 'FEMALE': px.colors.qualitative.Pastel[1]} # Assigna colors
)
fig_job_gender.update_layout(
    xaxis_title="Sector Laboral",
    yaxis_title="Nombre de Clients",
    xaxis_tickangle=-45
)
fig_job_gender.show() # Mostra el gràfic

# 4. Comprovació de valors nuls per gènere (en variables on n'hi ha)
print("\n\n4. Percentatge de Valors Nuls per Gènere en Columnes Crítiques:")
cols_with_nulls = ['LIMITE_CREDITO', 'IMPORTE_MINIMO_PAGOS']
null_analysis = df.groupby('GENDER')[cols_with_nulls].apply(lambda x: x.isnull().mean() * 100)
print(null_analysis)

print("\n" + "="*55 + "\nFi de l'Anàlisi Preliminar d'Imbalance de Gènere (amb Plotly).")
```
